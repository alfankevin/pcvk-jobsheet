5. Lakukan Face Tracking menggunakan Google Colab.

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load the pre-trained face cascade
cascade_wajah = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_alt.xml')

# Open a video capture object (0 corresponds to the default camera)
cap = cv.VideoCapture(0)  # Use cv.VideoCapture('your_video_file.mp4') for a video file

while True:
    # Read a frame from the video capture
    ret, frame = cap.read()

    if not ret:
        break  # Break the loop if the video capture fails

    # Convert the frame to grayscale for face detection
    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    # Detect faces in the frame
    roi_wajah = cascade_wajah.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Draw rectangles around the detected faces
    for (x, y, w, h) in roi_wajah:
        cv.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 3)

    # Display the frame with face tracking
    cv.imshow('Face Tracking', frame)

    # Break the loop if the 'q' key is pressed
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
cap.release()
cv.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
import pytesseract

ktp = cv.imread('/content/drive/MyDrive/ktp/ktp2.jpg')

gray_image = cv2.cvtColor(ktp, cv2.COLOR_BGR2GRAY)

_, threshold_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

detected_text = pytesseract.image_to_string(threshold_image)
print("Detected Text:\n", detected_text)

plt.imshow(threshold_image, cmap='gray')
plt.title('Detected Text')
plt.axis('off')
plt.show()
```
